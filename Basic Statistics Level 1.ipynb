# Q6) Calculate the Expected number of candies for a randomly selected child 
#Below are the probabilities of count of candies for children (ignoring the nature of the child-Generalized view)
#CHILD	Candies count	Probability
#A	1	0.015
#B	4	0.20
#C	3	0.65
#D	5	0.005
#E	6	0.01
#F	2	0.120
#Child A – probability of having 1 candy = 0.015.
#Child B – probability of having 4 candies = 0.20


```python
def expected_value(values, weights):
    values = np.asarray(values)
    weights = np.asarray(weights)
    return (values * weights).sum()/weights.sum()
```

```python
c_count = [1,4,3,5,6,2]
ch_prob = [0.015,0.20,0.65,0.005,0.01,0.120]
expected_value(c_count, ch_prob)
```

# 21) Check whether the data follows normal distribution
a)	Check whether the MPG of Cars follows Normal Distribution 
        Dataset: Cars.csv
b)	Check Whether the Adipose Tissue (AT) and Waist Circumference (Waist) from wc-at data set follows Normal Distribution 
       Dataset: wc-at.csv


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Load the dataset
file_path="C:\\Users\\chakri\\Downloads\\Cars.csv"
df=pd.read_csv(file_path)
df

# Extract the MPG column
mpg = df['MPG']

# Visualize the distribution
plt.figure(figsize=(8, 6))
plt.hist(mpg, bins=20, edgecolor = 'black',color='purple', alpha=0.7)
plt.title("Distribution of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()

# Shapiro-Wilk test for normality
stat, p_value = shapiro(mpg)

# Display the test result
print("Shapiro-Wilk Test for Normality:")
print("Statistic:", stat)
print("P-value:", p_value)
if p_value > 0.05:
    print("The data appears to be normally distributed.")
else:
    print("The data does not appear to be normally distributed.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Load the dataset
file_path="C:\\Users\\chakri\\Downloads\\wc-at.csv"
df=pd.read_csv(file_path)

# Extract the AT and Waist columns
at = df['AT']
waist = df['Waist']

# Visualize the distributions
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(at, bins=20, edgecolor='black', color='green', alpha=0.7)
plt.title("Distribution of Adipose Tissue (AT)")
plt.xlabel("AT")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(waist, bins=20, edgecolor='black',color='orange', alpha=0.7)
plt.title("Distribution of Waist Circumference (Waist)")
plt.xlabel("Waist")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Shapiro-Wilk test for normality - AT
stat_at, p_value_at = shapiro(at)

# Shapiro-Wilk test for normality - Waist
stat_waist, p_value_waist = shapiro(waist)

# Display the test results
print("\nShapiro-Wilk Test for Normality - Adipose Tissue (AT):")
print("Statistic:", stat_at)
print("P-value:", p_value_at)
if p_value_at > 0.05:
    print("The data for Adipose Tissue (AT) appears to be normally distributed.")
else:
    print("The data for Adipose Tissue (AT) does not appear to be normally distributed.")

print("\nShapiro-Wilk Test for Normality - Waist Circumference (Waist):")
print("Statistic:", stat_waist)
print("P-value:", p_value_waist)
if p_value_waist > 0.05:
    print("The data for Waist Circumference (Waist) appears to be normally distributed.")
else:
    print("The data for Waist Circumference (Waist) does not appear to be normally distributed.")
```

# Q 22) Calculate the Z scores of 90% confidence interval,94% confidence interval, 60% confidence interval 

```python
from scipy.stats import norm

confidence_intervals = [0.90, 0.94, 0.60]

z_scores = [norm.ppf((1 + ci) / 2) for ci in confidence_intervals]

for ci, z in zip(confidence_intervals, z_scores):
    print(f"Z-Score for {ci * 100}% Confidence Interval: {z}")
```

# Q 23) Calculate the t scores of 95% confidence interval, 96% confidence interval, 99% confidence interval for sample size of 25

```python
from scipy.stats import t

confidence_intervals = [0.95, 0.96, 0.99]
degrees_of_freedom = 25 - 1

t_scores = [t.ppf((1 + ci) / 2, df=degrees_of_freedom) for ci in confidence_intervals]

for ci, t_score in zip(confidence_intervals, t_scores):
    print(f"t-Score for {ci * 100}% Confidence Interval: {t_score}")
```

# Q 24)   A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days
Hint:  
   rcode   pt(tscore,df)  
 df  degrees of freedom


```python
from scipy.stats import t

sample_mean = 260
population_mean = 270
sample_std_dev = 90
sample_size = 18

df = sample_size - 1

t_score = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

probability = t.cdf(t_score, df)

print("Probability that 18 randomly selected bulbs would have an average life of no more than 260 days:", probability)
```
